```python
!pip install pyomo
from pyomo.environ import *
!apt-get install -y -qq glpk-utils
opt = SolverFactory('glpk', executable='/usr/bin/glpsol')

N = 10
model = AbstractModel()
model.i = RangeSet(N)
model.X = Var(model.i,model.i, bounds=(0,1), within=Reals)

def rule_C1(model):
  return sum (model.X[i,j] for i in model.i for j in model.i if i!=j) == 3
model.C1=Constraint(rule=rule_C1)


def rule_OF(model):
    return sum (model.X[i,j] for i in model.i for j in model.i if i!=j)
model.obj1 = Objective(rule=rule_OF, sense=minimize)
instance = model.create_instance()
results = opt.solve(instance)
instance.display()
```
